```python
## try,excpet,else block
try:
    num=int(input("Enter a number"))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter denominotor greter than 0")
except Exception as ex1:
    print(ex1)
else:
    print(f"The result is '{result}")
```

```python
## try ,except else and finally

try:
    num=int(input("Enter a number"))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter denominotor greter than 0")
except Exception as ex1:
    print(ex1)
else:
    print(f"The result is '{result}'")
finally:
    print("Excuetion Complete")

```
