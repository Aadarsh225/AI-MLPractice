```python
class Error(Exception):
    """Base class for all custom exceptions in this module."""
    pass
class dobException(Error):
    pass
```

```python
year = int(input("Enter your birth year: "))
age=2024 - year
try:
    if age <=30 and age>=20:
        print("Your age is valid and you can apply for exam")
    else:
        raise dobException
except dobException:
    print("Your age is not valid to apply for exam")
```
