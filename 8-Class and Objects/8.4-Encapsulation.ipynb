```python
### Encapsulation and Abstraction
## Encapsulation is the bundling of data and methods that operate on that data within a single unit or class.
## Encapsulation with getters and setters
# Punlic ,protected and private access modifiers
class Person:
    def __init__(self, name, age):
        self.name = name          # Public attribute - accessible from anywhere
        self._age = age          # Protected attribute - accessible within the class and its subclasses
        self.__ssn = "123-45-6789"  # Private attribute - accessible only within the class

    # Getter for private attribute
    def get_ssn(self):
        return self.__ssn

    # Setter for private attribute
    def set_ssn(self, ssn):
        self.__ssn = ssn

def get_person_info(person):
    return f"Name: {person.name}, Age: {person._age}"
person1 = Person("Alice", 30)
print(person1.name)          # Accessing public attribute
print(person1._age)         # Accessing protected attribute (not recommended)
print(person1.get_ssn())    # Accessing private attribute via getter

class ram(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id  # New attribute for subclass    
    def get_employee_info(self):
        return f"Name: {self.name}, Age: {self._age}, Employee ID: {self.employee_id}"
emp1 = ram("Ram", 28, "E123")
print(emp1.get_employee_info())  # Accessing inherited and new attributes
def get_name(person):
    return person.name
print(get_name(person1))  # Using function to access public attribute    
```
