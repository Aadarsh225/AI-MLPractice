```python
### PolyMorphism
## Method Overriding
## Base Class
class Animal:
    def speak(self):
        print("The animal makes a sound")
## Derived Class
class Dog(Animal):
    def speak(self):
        print("The dog barks")

class Cat(Animal):
    def speak(self):
        print("The cat meows")

## function that demonstates polymorphism
def animal_sound(animal):
    print(animal.speak())

dog1 = Dog()
dog1.speak()  # Output: The dog barks
cat1 = Cat()
cat1.speak()  # Output: The cat meows
animal_sound(dog1)  # Output: The dog barks
animal_sound(cat1)  # Output: The cat meows
```

```python
### Polymorphism with function and methods
## base class
class Shape:
    def area(self):
        return "The are of the figure is calculated"

## derived class 1
class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side * self.side
## derived class 2
class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length * self.breadth


## function that demonstrates polymorphism
def calculate_area(shape):
    print("Area:", shape.area())

sq1 = Square(4)
print(sq1.area())  # Output: 16
rect1 = Rectangle(4, 6)
print(rect1.area())  # Output: 24
calculate_area(sq1)  # Output: Area: 16
calculate_area(rect1)  # Output: Area: 24
```

```python
## Polymorphism with Abstract Base Classes
## Abstract Base Class are used to define common method for grouping of related object
from abc import ABC, abstractmethod

## Define Abstract Base Class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass
## Derived Class 1
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started with a key")
## Derived Class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine started with a button")

car1 = Car()
car1.start_engine()  # Output: Car engine started with a key    
motorcycle1 = Motorcycle()
motorcycle1.start_engine()  # Output: Motorcycle engine started with a button
```
