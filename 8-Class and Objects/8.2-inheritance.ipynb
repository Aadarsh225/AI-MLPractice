```python
### Inheritance(Single Level Inheritance)
## Parent Class
class Car:
    def __init__(self, windows, doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    def drive(self):
         print(f"The person will drive the {self.enginetype} car")
        
```

```python
car1 = Car(4, 4, "Petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, enginetype, autopilot):
        super().__init__(windows, doors, enginetype)
        self.autopilot = autopilot
    def enable_autopilot(self):
        if self.autopilot:
            print("Autopilot is enabled.")
        else:
            print("This car does not have autopilot feature.")
tesla1 = Tesla(4, 4, "Electric", True)
tesla1.drive() 
tesla1.enable_autopilot()    
```

```python
## Multi-Level Inheritance
## When a class inherits from more than one class
## base class 1
class Animal:
    def __init__(self,name):
        self.name = name
    def speak(self):    
        print("subclass must implement this method")
## Base class 2
class Pet():
    def __init__(self,owner):
        self.owner =owner

## Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner,breed):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
        self.breed = breed
    def speak(self):
        print(f"{self.name} says Woof! Woof!")
dog1 = Dog("Buddy","Alice","Golden Retriever")
dog1.speak()
print(f"{dog1.name} is owned by {dog1.owner} and is a {dog1.breed}.")


```
