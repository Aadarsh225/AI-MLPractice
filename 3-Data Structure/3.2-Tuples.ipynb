```python
## Acessing tuple elements
print(mixed_tuple[1])
print(mixed_tuple[-1])
print(mixed_tuple[1:3])
print(mixed_tuple[:3])
```

```python
## Immutable nature of tuples
## Tuples are immutable, meaning their elements cannot be changed after creation.

lst=[1,2,3,4,5]
print(lst)
lst[0]="aadarsh"
print(lst)
```

```python
numbers[1]="aadarsh"
```

```python
## Tuples Methods
print(numbers.count(3))
print(numbers.index(4))
```

```python
## packing and unpacking tuples
packed_tuple =1,"hello",3.4,True
print(packed_tuple)
```

```python
##unpacking tuples
a,b,c,d=packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
## unpacking with asterisk (*) operator
numbers=(1,2,3,4,5,6)
first, *middle, last=numbers
print(first)
print(middle)
print(last)
```

```python
## nested tuple

## nested list
lst=[[1,2],[3,4],[5,6]]
print(lst[1][0])

```

```python
nested_tuple=((1,2),(3,4),(5,6))
print(nested_tuple[1][0])
```

```python
## iterating through tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item)
```
