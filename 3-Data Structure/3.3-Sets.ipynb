```python
## Creation of Sets
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set([1,2,3,5,4,5,6])
print(my_empty_set)
print(type(my_empty_set))
```

```python
## Basic Operations on Sets
## Adding and removing elements
my_set.add(6)
print(my_set)
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(4)
print(my_set)
```

```python
my_set.remove(10)  # This will raise a KeyError since 10 is not in the set
```

```python
my_set.discard(10)  # No error if element not found
print(my_set)
```

```python
## pop method
popped_element = my_set.pop()
print("Popped element:", popped_element)
print("Set after pop:", my_set)
```

```python
## clear all the elements
my_set.clear()
print(my_set)
```

```python
## Sets membership test
my_set={1,2,3,4,5}
print(3 in my_set)  # True
print(10 in my_set) # False
```

```python
## Mathematical Set Operations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# Union
union_set=set1.union(set2)
print("Union:", union_set)

# Intersection
intersection_set=set1.intersection(set2)
print("Intersection:", intersection_set)

set1.intersection_update(set2)
print("Set1 after intersection_update with Set2:", set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
# Difference
difference_set=set1.difference(set2)
print("Difference (Set1 - Set2):", difference_set)
```

```python
set2.difference(set1)
print("Difference (Set2 - Set1):", set2.difference(set1))
```

```python
## symmetric difference
set1.symmetric_difference(set2)
```

```python
## Sets Method
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print(set1.issubset(set2))

set1.issuperset(set2)
```

```python
lst=[1,2,2,3,4,4,5]
set(lst)
```

```python
### counting unique word in text

text="In this tutorial we are discussing about sets"
words=text.split()

##convert list of word into unique words
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
