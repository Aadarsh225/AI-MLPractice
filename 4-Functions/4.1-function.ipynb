```python
## whys functions

number = 10
if number % 2 == 0:
    print(f"{number} is even")  
else:
    print(f"{number} is odd")
```

```python
def check_even_odd(number):
    """Check if a number is even or odd."""
    if number % 2 == 0:
        return f"{number} is even"
    else:
        return f"{number} is odd"
```

```python
check_even_odd(10)
```

```python
## function with multiple parameters
def add_numbers(a,b):
    """Return the sum of two numbers."""
    return a + b
```

```python
#default parameters
def greet(name, greeting="Hello"):
    """Greet a person with a given greeting."""
    return f"{greeting}, {name}!"
greet("Alice")
greet("Bob", "Hi")
```

```python
### Varible length arguments
## positional and keyword arguments
## positional variable length arguments
def print_number(*args):
    """Print all provided numbers."""
    for number in args:
        print(number)
```

```python
print_number(1, 2, 3, 4, 5)
```

```python
## keyword variable length arguments
def print_details( **kwargs):
    """Print all provided key-value pairs."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Alice", age=30, city="New York")
```

```python
## positional and keyword arguments
def print_Details(*args, **kwargs):
    """Print positional and keyword arguments."""
    print("Positional arguments:")
    for arg in args:
        print(arg)
    print("Keyword arguments:")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_Details(1, 2, 3, name="Alice", age=30)
```
