```python
## syntax
lambda arguments: expression
```

```python
def add(x, y):
    return x + y
```

```python
add(2, 3)
```

```python
addition=lambda x, y: x + y
type(addition)
print(addition(2, 3))
```

```python
def even(num):
    if num % 2 == 0:
        return True
    else:
        return False    
    
even(4)
```

```python
even_lambda = lambda num: True if num % 2 == 0 else False
even_lambda(4)
```

```python
def addition(x, y,z):

    return x + y + z
addition(2, 3, 4)
```

```python
addition1 = lambda x, y, z: x + y + z
addition1(2, 3, 4)
```

```python
## map() with lambda : applies a function to all the items in an input list.
numbers = [1, 2, 3, 4, 5]
def square(num):
    return num ** 2

square(2)
```

```python
lambda_squared = list(map(lambda num: num ** 2, numbers))
lambda_squared
```
